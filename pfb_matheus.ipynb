```python
ntap = 4
samples = 2000
t = np.linspace(0,samples*ntap,samples*ntap)
f1 = 400
f2 = 401
f3 = 402
f4 = 403
fun = np.sin(2*np.pi*f1*t) + np.sin(2*np.pi*f2*t) +np.sin(2*np.pi*f3*t)+np.sin(2*np.pi*f4*t)+np.random.randn(t.size)

plt.plot(fun)
plt.title('Time stream')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

plt.plot(np.fft.fft(fun))
plt.title('Fourier transform')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()
```

```python
x2 = np.linspace(-ntap,ntap,ntap*samples)
sinc = np.sinc(x2)
plt.plot(sinc)
plt.title('Sinc function')
plt.show()

plt.plot(np.fft.fft(sinc))
plt.show()

fun *= sinc

plt.plot(fun)
plt.title('Signal * sinc')
plt.show()
```

```python
def split(x,ntap):
    a = np.split(x,ntap)
    sum(a)
    return a
def soma(x):
    b = sum(x)
    return b 
a = soma(split(fun,ntap))

plt.plot(a)
plt.show()

plt.plot(np.fft.fft(a))
plt.show()
```

```python
plt.figure(figsize=(10,5))
# Or equivalently,  "plt.tight_layout()"

plt.subplot(1, 2, 1)
plt.specgram(fun, Fs=2, scale = 'linear')
plt.title('Spectrogram of f')
plt.xlabel('Time(s)')
plt.ylabel('Frequency(Hz)')
plt.tight_layout()
plt.colorbar()

plt.subplot(1, 2, 2)
plt.specgram(a,Fs=2, scale = 'linear')
plt.title('Spectrogram with 4-tap PFB')
plt.xlabel('Time(s)')
plt.ylabel('Frequency(Hz)')
plt.tight_layout()
plt.colorbar()
plt.savefig('comparison4tapnoise6.png')
plt.show()
```
